```python
import yfinance as yf
import statsmodels.api as sm
from statsmodels import regression
import numpy as np
import matplotlib.pyplot as plt

# Define the linear regression function
def linreg(X, Y):
    # Running the linear regression
    X = sm.add_constant(X)
    model = regression.linear_model.OLS(Y, X).fit()
    a = model.params[0]
    b = model.params[1]
    X = X[:, 1]

    # Return summary of the regression and plot results
    X2 = np.linspace(X.min(), X.max(), 100)
    Y_hat = X2 * b + a
    plt.scatter(X, Y, alpha=0.3)  # Plot the raw data
    plt.plot(X2, Y_hat, 'r', alpha=0.9)  # Add the regression line, colored in red
    plt.xlabel('X Value')
    plt.ylabel('Y Value')
    plt.show()

    return model.summary()

# Define the start and end dates for the data
start_date = '2020-01-01'
end_date = '2023-01-01'

# List of 50 well-known stocks
stock_list = ['AAPL', 'GOOGL', 'AMZN', 'MSFT', 'TSLA', 'FB', 'NFLX', 'NVDA', 'INTC', 'AMD',
              'BA', 'BABA', 'BRK-B', 'JPM', 'V', 'WMT', 'DIS', 'PYPL', 'MA', 'HD',
              'PFE', 'MRNA', 'XOM', 'CVX', 'KO', 'PEP', 'NKE', 'T', 'VZ', 'WFC',
              'UNH', 'JNJ', 'PG', 'ADBE', 'CRM', 'CSCO', 'ORCL', 'IBM', 'QCOM', 'MMM',
              'GE', 'CAT', 'MCD', 'ABBV', 'TMO', 'UPS', 'FDX', 'SBUX', 'GS', 'SPGI']

# Get the benchmark data (SPY) using yfinance
benchmark = yf.download('SPY', start=start_date, end=end_date)['Adj Close']
r_b = benchmark.pct_change()[1:]  # Get the benchmark returns

# Loop through each stock in the list, perform linear regression, and display the summary
for stock in stock_list:
    try:
        # Get the asset data using yfinance
        asset = yf.download(stock, start=start_date, end=end_date)['Adj Close']
        r_a = asset.pct_change()[1:]  # Get the asset returns

        print(f"\nLinear Regression for {stock} against SPY:")
        summary = linreg(r_b.values, r_a.values)
        print(summary)

    except Exception as e:
        print(f"Could not process stock {stock} due to error: {e}")
```

```python
import yfinance as yf
import statsmodels.api as sm
from statsmodels import regression
import numpy as np
import matplotlib.pyplot as plt

# Define the linear regression function
def linreg(X, Y):
    # Running the linear regression
    X = sm.add_constant(X)
    model = regression.linear_model.OLS(Y, X).fit()
    a = model.params[0]
    b = model.params[1]
    X = X[:, 1]

    # Return summary of the regression and plot results
    X2 = np.linspace(X.min(), X.max(), 100)
    Y_hat = X2 * b + a
    plt.scatter(X, Y, alpha=0.3)  # Plot the raw data
    plt.plot(X2, Y_hat, 'r', alpha=0.9)  # Add the regression line, colored in red
    plt.xlabel('X Value')
    plt.ylabel('Y Value')
    plt.show()

    return model.summary()

# Define the start and end dates for the data
start_date = '2020-01-01'
end_date = '2023-01-01'

# List of stocks outside S&P 500, mentioned earlier
stock_list = [
    'DKNG',  # DraftKings - USD (U.S. market)
    'CPNG',  # Coupang - USD (U.S. market)
    'ROKU',  # Roku - USD (U.S. market)
    'PLUG',  # Plug Power - USD (U.S. market)
    'TWLO',  # Twilio - USD (U.S. market)
    'SHOP.TO',  # Shopify - CAD (Canada market)
    'MELI',  # MercadoLibre - USD (U.S. market)
    'UPST',  # Upstart - USD (U.S. market)
    'PLTR',  # Palantir - USD (U.S. market)
    'U',     # Unity Software - USD (U.S. market)
    'AFRM'   # Affirm - USD (U.S. market)
]

# Get the benchmark data (SPY) using yfinance (in USD)
benchmark = yf.download('SPY', start=start_date, end=end_date)['Adj Close']
r_b = benchmark.pct_change()[1:]  # Get the benchmark returns

# Placeholder for currency conversion if necessary (for example CAD to USD for Shopify)
# You can use forex-python or another API to fetch the conversion rate
conversion_rates = {
    'CAD': 0.75,  # Example rate for CAD to USD
    'USD': 1.00   # No conversion needed for USD stocks
}

# Loop through each stock, perform linear regression, and display the summary
for stock in stock_list:
    try:
        # Get the asset data using yfinance
        asset = yf.download(stock, start=start_date, end=end_date)['Adj Close']
        r_a = asset.pct_change()[1:]  # Get the asset returns

        # Check if the stock is listed outside the U.S. and convert the currency if needed
        currency = yf.Ticker(stock).info.get('currency', 'USD')
        conversion_rate = conversion_rates.get(currency, 1.0)
        r_a_converted = r_a * conversion_rate  # Convert returns to USD equivalent

        print(f"\nLinear Regression for {stock} (currency: {currency}) against SPY:")
        summary = linreg(r_b.values, r_a_converted.values)
        print(summary)

    except Exception as e:
        print(f"Could not process stock {stock} due to error: {e}")
```
